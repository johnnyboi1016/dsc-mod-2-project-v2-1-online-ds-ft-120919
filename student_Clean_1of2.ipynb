## Data Cleaning

#### Drop the 'id' column (make it the index), check for duplicates, null values

```python
df = raw_data.set_index('id')
#Check column data types, # null values per column, # of row duplicates and # of index (id) duplicates
display(df.info(), df.isna().sum(), f'Duplicates: {df.duplicated().sum()}', f'ID Duplicates: {df.index.duplicated().sum()}')
```

#### Let's take a closer look at the 177 index 'id' duplicates - likely subsequent sales

```python
#retain the 1st instance for duplicates; df_dup will store all of our properties that have been sold more than once
df_dup = df[df.index.duplicated(keep=False)]
df_dup = df_dup.sort_values(by='id')
df_dup
```

### Confirmed - same property ids show a different date/price. Now let's clean up the 'waterfront' column
* Note the null waterfront value paired with a 0 for the same property.. before we replace all of them with 0 let's confirm none are paired with a 1

```python
#first locate index numbers for null waterfront values then filter for waterfront values of 1 matching those index numbers,
#then print total sum of matches
np.sum(df_dup.loc[df_dup[df_dup.waterfront.isna()].index].waterfront==1)
```

```python
#None found so replacing waterfront nulls with 0.0 should be appropriate
df.waterfront.fillna(0.0, inplace=True)
df_dup.waterfront.fillna(0.0, inplace=True)
print(df.waterfront.value_counts(),'\n',f'Null values: {df.waterfront.isna().sum()}')
```

#### That's not too many waterfront properties! Let's take a closer look.

```python
#Get mean/median/std of both groups and print average price differences
print('Waterfront Properties:', f'Mean: {df[df.waterfront==1].price.mean().round()}',
      f'Median: {df[df.waterfront==1].price.median().round()}', f'Std: {df[df.waterfront==1].price.std().round()}')
print('Non-waterfront Properties:', f'Mean: {df[df.waterfront==0].price.mean().round()}',
      f'Median: {df[df.waterfront==0].price.median().round()}', f'Std: {df[df.waterfront==0].price.std().round()}')
print('Difference: ', f'Mean: {(df[df.waterfront==1].price.mean().round() - df[df.waterfront==0].price.mean().round())}',
      f'Median: {(df[df.waterfront==1].price.median().round() - df[df.waterfront==0].price.median().round())}')
```

#### Despite these 2 groups having a fairly consistent average price difference ($1.06-1.18M), the significantly higher standard deviation (1.14M vs 342k) and extremely small sample size make these properties good candidates to help reduce the variance in our predictive model (some could consider them as outliers). The 146 waterfront properties in this dataset make up less than 0.7% of our dataset so we will go ahead and remove those properties (and the feature column).

```python
#Store these properties in df_water to revisit later
df_water = df[df.waterfront==1].copy()
df = df.drop(df_water.index)
df = df.drop(labels='waterfront', axis=1)
```

### Now let's examine the 'view' column to determine the best cleaning method

```python
df.view.value_counts(dropna=False)
```

#### This column described as 'has been viewed' with a range of values from 0-4 for a 20k+ dataset do not make sense as a 'running count' of how many times this property has been viewed by interested buyers before being sold. It actually appears to be a rating of 'how good the view is' from the property (on a scale of 0-4). Let's take a closer look.

```python
#Show duplicate property sales side by side to see if view count changes
df_dup[df_dup.view>0]
```

#### It is very unlikely that the same property would be viewed the exact same number of times when it gets sold more than once. Therefore we will treat this feature similarly to grade / condition (overall grade given to the housing unit, based on King County grading system / how good the condition is overall).
* This feature may not be so useful with so many zero values..

```python
#Filling in a 0 for null views
df.view.fillna(0.0, inplace=True)
print(df.view.value_counts(dropna=False, normalize=True))

#View histogram
df.view.hist(bins=30)
```

#### Considering that over 90% of our dataset has a view of 0 and how non-normal its distribution is, we will just drop this feature column.

```python
df = df.drop('view',axis=1)
```

### Moving onto the 'yr_renovated' column

```python
#First replace null values with 0
df.yr_renovated.fillna(0.0, inplace=True)
print(f'Null values: {df.yr_renovated.isna().sum()}')
df.yr_renovated.value_counts(normalize=True)
```

#### 0/null values would indicate no renovation having been done since the original year built. Considering these represent over 96% of our dataset, we will drop this column altogether. It may be useful to have a renovation (yes/no) column or 'recent renovation' which we can explore later.

```python
df.drop('yr_renovated', axis=1, inplace=True)
```

### Let's find out why the 'sqft_basement' column type is listed as 'object'

```python
df.sqft_basement.value_counts()
```

#### Determine the best method for cleaning up the '?' values

```python
#First check for '?' in our duplicated set to see if we have proper values we can use to fill
df_dup.loc[df_dup[df_dup.sqft_basement=='?'].index]
```

#### The last property confirms that we should be filling in the unknown sqft_basement values with the difference between sq_living and sqft_above. Let's see if we can do this with a single line of code!

```python
#Return same numerical value or replace the '?' with the proper difference between living and above
df.sqft_basement = df.apply(lambda x: x.sqft_basement if x.sqft_basement != '?' else (x.sqft_living - x.sqft_above), axis=1)

#Now should be able to convert whole column to correct type - had to do float first so the '0.0's did not create an error
df.sqft_basement = df.sqft_basement.astype('float64').astype('int64')
```

```python
#Now let's double check that (sqft_above + sqft_basement = sqft_living) for every property
x = 0   #counter
for s in df.sqft_living:
    if s != (df.sqft_above.iloc[x] + df.sqft_basement.iloc[x]):
        print(df.sqft_living.iloc[x]) #print discrepancies; no print output confirms none
    x+=1
```

## Dropping outliers
We will stick with a general rule of attempting to stay within ~3-4 standard deviations from the mean, preserving normality and as much of our original dataset as possible. It is always easier to drop more outliers later when testing our regression model to improve results.

```python
#Examining price outliers
print(df.price.describe().round(),'\n',f'99th percentile: {df.price.quantile(q=.99)}')
print('Mean plus 3 stds: ', df.price.mean() + (df.price.std() * 3))
```

#### Interesting.. the cutoff at 3 standard deviations above the mean should account for ~99.7% of the data and be greater than the 99th percentile value (if normally distributed). Let's take a closer look..

```python
#See how many values would be dropped and visually compare differences
print(df[df.price>1560710].shape, df[df.price>1870000].shape)
df[df.price<1870000].price.hist(), df[df.price<1560710].price.hist();
```

#### A difference of <200 and both represent <1-2% of our dataset; considering how heavily right tailed this distribution is we'll go with the lower mean+3std cutoff

```python
df.drop(df[df.price>1560710].index, inplace=True)

#Get an idea of the other features/variables
df.describe().round()
```

#### When comparing the 75% quantile and max values, a few features stand out. Let's first take a closer look at: sqft_lot, 15 and sqft_basement

```python
#Simple function to print a given column's mean + 3 standard deviations and the top 10 values sorted by given column, along
#with its distribution histogram and scatterplot vs price with prediction line
def outliers(col):
    print(col.upper())
    print(f'Mean + 3std: {df[col].mean() + (df[col].std() * 3)}')
    print(df[col].value_counts(normalize=True).head())
    display(df.sort_values(by=col, ascending=False).head(10))
    df[col].hist(bins=30)
    sns.lmplot(x=col, y='price', data=df, line_kws={'color': 'r'})
    #plt.title("{} vs {}".format(col, target))
    plt.show();

outliers('sqft_lot'), outliers('sqft_basement'), outliers('sqft_lot15')
```

#### Talk about heavy positive skew all around! We'll start with these initial outliers:
* Sqft_lot: > 650k (9 properties)
* Sqft_basement: none for now
* Sqft_lot15: >= 400k (6)

```python
df = df.drop(df[df.sqft_lot>650000].index)
df = df.drop(df[df.sqft_lot15>=400000].index)
```

#### Now let's take a look at the other features: bedrooms, bathrooms, sqft_living, floors, waterfront, view, condition, grade, sqft_above, yr_built, zipcode, lat, long, sqft_living15

```python
for col in df.drop(['date','price','sqft_lot','sqft_lot15','sqft_basement'],axis=1).columns:
    outliers(col)
```

#### Dropped 12 outliers total:
* Bedrooms: > 10 (2 properties)
* Bathrooms: > 7 (1)
* Sqft_living: > 7000 (4)

```python
df = df.drop(df[df.bedrooms>10].index)
df = df.drop(df[df.bathrooms>7].index)
```

3. Dropped the yr_renovated and view column.
4. Dropped over 400 price and feature outliers.
